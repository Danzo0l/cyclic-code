```python
import random
import matplotlib.pyplot as plt

# Number of messages
MESSAGES = 1000
# Param of repeats
REPEAT = 1000
# Time parametr
TAO = 0.5
# Accuracy of channel
ACCURACY = 0.1

Perr = [(i * ACCURACY) for i in range(int(1 / ACCURACY) + 1)]
mu = []
# Simulation of algorhytm
for l in Perr:
    t = 0
    for i in range(MESSAGES):
        # Send message
        # print(f"send: t: {t}")
        t += 1
        k = False
        count = 0
        while (k != True):
            # print(f"\treception: t: {t}, k: {k}", end=';\n')
            i_hat = random.random()
            t += TAO
            if (i_hat > l):
                k = True
                break
            if (count > REPEAT):
                break
            i_hat_rev = random.random()
            if (i_hat_rev <= l):
                k = False
            count += 1
    mu.append(MESSAGES/t)
    print(f"recive: Perr: {round(l, 2)}, t: {t}, N: {i + 1}, k: {k}")

print(mu)
plt.plot(Perr, mu)
plt.xlabel('Perr')
plt.ylabel('mu')
plt.title('Graph of channel mu')
plt.grid(True)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

def send_packet(p_error):
    # Функция отправки пакета с вероятностью ошибки p_error
    if random.random() > p_error:
        return True  # Пакет успешно отправлен
    else:
        return False  # Пакет потерян из-за ошибки в канале

def receive_ack(delay):
    # Функция приема квитанции с задержкой delay
    pass  # В данном примере просто заглушка

def arq_algorithm(p_error, delay):
    t = 1 / p_error  # Рассчитываем t как обратную величину от p_error
    # Отправляем пакет и ждем квитанции
    packet_sent = False
    while not packet_sent:
        packet_sent = send_packet(p_error)
    
    receive_ack(delay)
    return t

# Задаем значения для p_error и delay
p_error_values = [(i * ACCURACY) for i in range(1, int(1 / ACCURACY) + 1)]
delay = 2

# Создаем списки для хранения значений 1/t и p_error
inv_t_values = []
p_error_values_plt = []

# Выполняем алгоритм для каждого значения p_error
for p_error in p_error_values:
    t = arq_algorithm(p_error, delay)
    inv_t_values.append(1 / t)
    p_error_values_plt.append(p_error)

# Строим график зависимости 1/t от p_error
plt.plot(p_error_values_plt, inv_t_values, marker='o')
plt.xlabel('Perr')  # Подпись оси x
plt.ylabel('1/t')   # Подпись оси y
plt.title('Зависимость 1/t от Perr')  # Заголовок графика
plt.grid(True)      # Включаем сетку
plt.show()          # Показываем график
```
